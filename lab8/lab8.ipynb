## EOQ

Модель EOQ описує, як визначити оптимальний розмір замовлення, щоб мінімізувати загальні витрати.

**Основні припущення та обмеження:**

- Попит D відомий і постійний протягом року.
- Вартість замовлення S та зберігання H — сталі.
- Постачання миттєве, відсутні затримки.
- Не враховуються зовнішні фактори такі як економічні коливання, дефіцит, інфляція тощо.

## Багаторівнева система
Інтеграція моделей різного рівня деталізації + мультидисциплінарний підхід.

### Додатковий рівень:
- Макрорівень – модель економічного середовища (інфляція, зміни у вартості S і H, зовнішній попит).
- Мікрорівень – класична EOQ.

## Модифікована EOQ

Ми враховуємо зміни з макрорівня у вигляді функцій часу або зовнішніх факторів:


#### 1. Вплив інфляції $H(t)$:

$$
H(t) = H_0 \cdot (1 + \pi)^t
$$

де:
- $H_0$ — початкова вартість зберігання,  
- $\pi$ — річна інфляція,  
- $t$ — кількість років.


#### 2. Сезонні коливання попиту:

$$
D(t) = D_0 \cdot \left( 1 + a \cdot \sin(\omega t + \phi) \right)
$$

де:
- $D_0$ — базовий попит,  
- $a$ — амплітуда сезонних змін ($0 < a < 1$),  
- $\omega$ — частота коливань,  
- $\phi$ — зсув фази.


#### 3. Зміна вартості замовлення:

$$
S(t) = S_0 \cdot \left(1 + \delta(t) \right)
$$

де:
- $S_0$ — базова вартість замовлення,  
- $\delta(t)$ — функція зовнішніх впливів:

$$
\delta(t) = \beta_1 \cdot f_{\text{пальне}}(t) + \beta_2 \cdot f_{\text{курс}}(t)
$$


#### 4. Модифікована формула EOQ:

$$
Q^*(t) = \sqrt{ \frac{2 \cdot D(t) \cdot S(t)}{H(t)} }
$$


#### 5. Загальні витрати з урахуванням змін:

$$
TC(Q, t) = \frac{D(t)}{Q} \cdot S(t) + \frac{Q}{2} \cdot H(t)
$$


Багатоетапність дозволяє комбінувати просту модель з ускладненням на основі реального середовища. Також гнучка структура моделі дозволяє використовувати її для стратегічного планування.

## Нові можливості моделі
- Адаптація до реального часу.
- Аналіз впливу інфляції, пального, валютного курсу.
- Можливість прогнозування витрат у динаміці.
- Можна вбудувати прогнози на основі машинного навчання.
